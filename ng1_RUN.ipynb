```c++
TH2D *qdceehist = new TH2D("qdceehist","qdceehist",500,0.5,30,500,0.2,12);
tree->Draw(Form("fast_ee[%d]:all_ee[%d]>>qdceehist",ch,ch));
qdceehist->Draw("colz");
c1->Draw();
```

ng弁別のhistgram

```c++
TH2D *funch = new TH2D("funch","funch",500,0.5,30,500,0.2,12);
tree->Draw(Form("function[%d]:all_ee[%d]>>funch",ch,ch)
           ,Form("all_ee[%d]>%f",ch,offset));
qdceehist->Draw("colz");
funch->SetMarkerColor(kRed);
funch->Draw("same");
qdceehist->SetStats(0);
//c1->Draw();
```

```c++
TFile *f = TFile::Open(Form("./../root/%04d_ng1.root",LiRUN));
TTree *tree = (TTree *)f->Get("tree");
TH2D *qdceehistLi = new TH2D("qdceehistLi","qdceehistLi",500,0.5,30,500,0.2,12);
tree->Draw(Form("fast_ee[%d]:all_ee[%d]>>qdceehistLi",ch,ch));
qdceehist->SetTitle(Form("C & Li %04d_%d;all[MeVee];fast[MeVee]",RUN,ch));
qdceehist->Draw("colz");
qdceehistLi->SetMarkerColor(kRed);
qdceehistLi->Draw("same");
c1->Draw();
//c1->SaveAs(Form("./../figure/C0313_%04d_%d.pdf",RUN,ch));  /////
```
